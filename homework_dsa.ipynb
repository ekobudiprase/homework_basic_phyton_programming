**Fixing the data**

```python
import numpy as np
import pandas as pd

data = pd.read_csv("googleplaystore.csv", encoding="latin1")
```

**Normalizing data**

```python
data.drop_duplicates(subset='App', inplace=True)
data['Installs'] = data['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
data['Installs'] = data['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)
data['Installs'] = data['Installs'].astype(int)
data['Rating'] = data['Rating'].astype(float)
data['Reviews'] = data['Reviews'].astype(int)
```

<br> **4. What are top 3 apps that have highest number of users installed/download among the following categories: (Hint:
use installs column)**

```python
fol_cat = ["BUSINESS", "COMMUNICATION", "GAME", "DATING", "MUSIC"]
data_fol_cat = data[data['Category'].isin(fol_cat)]
data_fol_cat.sort_values(['Category','Installs'],ascending=False).groupby('Category').head(3).filter(['App','Category','Installs'])
```

<br> **5. What are top 3 apps that have highest rating among the following categories**

```python
data_fol_cat.sort_values(['Category','Rating'],ascending=False).groupby('Category').head(3).filter(['App','Category','Rating'])
```
